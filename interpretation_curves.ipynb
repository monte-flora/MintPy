```python
# Import Modules
import sys,os
sys.path.append(os.getcwd())

from clarifier_plots import ClarifierPlot
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.linear_model import LogisticRegression

from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
import pandas as pd
import pickle
```

```python
# Load the breast cancer classification dataset available
# in the sklearn package
data = load_breast_cancer()
X, y, feature_names = data.data, data.target, data.feature_names
X_train, X_valid, y_train, y_valid = train_test_split(X, y, random_state=42)

# Train a random forest
rf= RandomForestClassifier(n_estimators=10, random_state=42)
rf.fit(X_train, y_train)

#Train logistic regression model
lr = LogisticRegression()
lr.fit(X_train, y_train)

#Train a gradient-boosted tree
gb = GradientBoostingClassifier()
gb.fit(X_train, y_train)
```

```python
c_plt = ClarifierPlot(models={'RandomForest': rf, 'LogisticRegression': lr, 'GradientBoosted': gb}, 
                      examples=X_train, targets=y_train, feature_names=feature_names)
```

```python
important_vars = ['compactness error', 'texture error', 'area error', 'mean smoothness']
fig, axes = c_plt.plot_pdp(important_vars)
```

```python
c_plt = ClarifierPlot(models=gb, examples=X_train, targets=y_train, feature_names=feature_names)
important_vars = ['compactness error', 'texture error', 'area error', 
                  'mean smoothness', 'worst compactness', 'worst compactness']
fig, axes = c_plt.plot_pdp(important_vars)

from matplotlib.patches import Patch
from matplotlib.lines import Line2D
legend_elements = [Line2D([0], [0], color='b', lw=4, label='Line')]

lax = axes.flat[-1]
#fig.legend(handles=legend_elements, loc=10)
h,l = axes.flat[-1].get_legend_handles_labels()
print(h,l)
lax.legend(h,l, borderaxespad=0)
lax.axis("off")
```
